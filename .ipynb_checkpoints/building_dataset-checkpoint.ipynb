```python
import spacy
import re
import pandas as pd
```

# Step 1

```python
with open("data.txt", 'r') as file:
    data = [line.strip() for line in file]
    
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
dataframe = list()
for i,abst in enumerate(data):
    doc = nlp(abst)
    for sent in doc.sents:
        for token in sent:
            dataframe.append([i,token.text])
            
df = pd.DataFrame(dataframe,columns =['id','token'])
df.to_csv('data.csv',index=False)
```

```python
len(df.token)
```

```python
tokens = set(df.token.to_list())
len(set(tokens))
```

```python
from functionwords import FunctionWords
```

```python
fw = FunctionWords(function_words_list='english')
fw = fw.get_feature_names()
len(fw)

tokens_wo_fw = [token for token in tokens if token.lower() not in fw]
len(tokens_wo_fw)
```

```python
import string 

punc = string.punctuation+"—“”"

tokens_wo_p = [token for token in tokens_wo_fw if token not in punc]


len(tokens_wo_p)
```

```python
import re

p = re.compile("^\d+,\d+|\d+\.\d+|\d+")

tokens_wo_nb = [token for token in tokens_wo_p if not p.fullmatch(token)]

len(tokens_wo_nb)
```

```python
with open('tokens.txt','w') as file:
    for token in tokens_wo_nb:
        file.write(token+"\n")
        
```

# Step 2

```python
with open('tokens.txt','r') as file:
    annotation = {}
    for line in file:
        line = line.strip()
        try:
            annotation[line.split()[0]] =  line.split()[1]
        except:
            annotation[line.split()[0]] = 'O'
        
annotation   
        
        
```

```python
def annotate(x):
    
    if x in annotation:
        return annotation[x]
    else:
        return 'O'
    
    
df['annotation'] = df['token'].apply(lambda x: annotate(x))
df
```

```python
df.to_csv('data.csv',index=False)
```

```python
for i in range(10):
    data = ann_df[ann_df.id==i]
    data = data[['token','annotation']]
    data = data.values.tolist()
    with open(f'annotations/article-{i+1}.ann1','w') as file:
        for row in data:
            line = " ".join(row)+"\n"
            file.write(line)
```
