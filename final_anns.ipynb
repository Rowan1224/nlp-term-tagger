```python
import pandas as pd
import re
```

```python
final_anns = pd.read_csv('2016-train.csv')
colRegex = re.compile(r"\d$|^diff")
```

```python
bad_cols = [col for col in final_anns.columns if colRegex.search(col)] 
print(bad_cols)
final_anns
```

```python
final_anns.drop(bad_cols, axis=1, inplace=True)
final_anns.dropna(inplace=True)
```

```python
final_anns
```

```python
final_anns.loc[(final_anns['article'] == "2016-42-3-391-419") | (final_anns['article'] == "2016-42-2-277–306")]
#final_anns.loc[(final_anns['article'] == "2016-42-3-391-419") & (final_anns['article'] == "2016-42-2-277–306")].any().all()
final_anns.loc[(final_anns['article'].isin(["2016-42-3-391-419", "2016-42-2-277–306"]))]
```

```python
articles = final_anns["article"].unique()
```

```python
for art in articles:
    filename = art + ".final"
    data = final_anns[final_anns['article'] == art]
    flattened = data[["token", "final"]].values.tolist()
    with open("2016_dir/" + filename, 'w') as file:
        for row in flattened:            
            line = " ".join(row) + "\n"
            file.write(line) 
    file.close()
    
```
